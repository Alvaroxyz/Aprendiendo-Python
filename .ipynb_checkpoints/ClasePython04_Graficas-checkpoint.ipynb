```python
#para graficar usamos np y plt
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline 
```

**Gráficando en Python**

```python
#Creamos una función que haga el cálculo de las imágenes
def f(x):
    return (x**2)*(np.sin(x))

#Definimos un dominio
x = np.arange(-100, 100, 0.1)
#arrange es un vector unidimensional: arange(start, stop, step) Values are generated within the
#half-open interval [start, stop), with spacing between values given by step.

#Creamos el vector de las imágenes
y = f(x)

#Mandamos a graficar 
plt.plot(x,y)
```

```python
plt.plot(x,y, label='Onda radio')
plt.title('Función contorno de $x^2$')
plt.xlabel('Variable $x$ cambiando')
plt.ylabel('Variable $y$ oscilando')
plt.grid()
plt.legend()
plt.show()
```
