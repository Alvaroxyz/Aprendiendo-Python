```python
diccionario={'Fizz':[],'Buzz':[],'FizzBuzz':[]}
a = 1
while (a<=100):
    if a%3==0:
        if a%5==0:
            print('FizzBuzz')
            diccionario['FizzBuzz'].append(a)
        else:
            print('Fizz')
            diccionario['Fizz'].append(a)
    elif a%5==0:
        print('Buzz')
        diccionario['Buzz'].append(a)
    else:
        print(a)
    a=a+1

print(diccionario)
```
